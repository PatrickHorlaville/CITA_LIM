```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import astropy.units as u # I use astropy units in everything, very helpful for bookkeeping/debugging

from lim import lim # The base intensity mapping object

matplotlib.rcParams.update({'font.size': 18,'figure.figsize':[8,7]}) # Make figures better
#matplotlib.rcParams,update({}'figure.figsize'] = [8,7]
from scipy.ndimage import gaussian_filter
```

```python
# Initialize LineModel with default model
m = lim()
m_cii = lim('Lichen_v2', doSim = True)
m_cii_old = lim('Lichen', doSim = True)
m_co = lim('TonyLi_PhI', doSim = True)
```

```python
m_cii.update(tobs = 40000*u.hr)
m_cii_old.update(tobs = 40000*u.hr)
```

## Comparing Power Spectra

```python
m_co.update(linear_k = True)
m_cii.update(linear_k = True)
m_cii_old.update(linear_k = True)
```

```python
k = m_cii.k
```

```python
Pk = m_cii.Pk
Pk_old = m_cii_old.Pk
```

```python
plt.plot(k, Pk, label = 'New Signal')
plt.plot(k, Pk_old, label = 'Old Signal')
plt.plot(m_co.k, m_co.Pk, label = 'CO')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('k')
plt.ylabel('P(k)')
plt.title(r'Old vs New Signal')
plt.legend()
plt.grid()
```

```python
plt.plot(k, Pk, '-o', label = 'New Signal')
plt.plot(k, Pk_old, '-o', label = 'Old Signal')
plt.plot(m_co.k, m_co.Pk, '-o', label = 'CO')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('k')
plt.ylabel('P(k)')
plt.title(r'Old vs New Signal')
plt.legend()
plt.grid()
```
